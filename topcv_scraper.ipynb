```python
import requests
from bs4 import BeautifulSoup
```

```python
TARGET_URL = 'https://www.topcv.vn/tim-viec-lam-moi-nhat-tai-ho-chi-minh-l2?exp=1&type_keyword=0&sba=1&locations=l2'

# Headers để giả lập trình duyệt
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9,vi;q=0.8', # Yêu cầu ngôn ngữ
    'Referer': 'https://www.topcv.vn/' # Trang mà từ đó ta "đến" trang này
}

print(f"URL muc tieu can crawl la: {TARGET_URL}")
```

```python
def fetch_page_content(url):
    """
    Hàm này sẽ gửi yêu cầu GET đến URL và trả về nội dung html dưới dạng text.
    """
    try:
        print(f"Dang gui request den: {url}")
        response = requests.get(url, headers=HEADERS, timeout=10)
        
        response.raise_for_status()  # Kiểm tra xem yêu cầu có thành công không (200 la OK)
        print(f"Request thanh cong! Status code la: {response.status_code}")
        
        return response.content #Trả về content (bytes) để BeautifulSoup tự xử lý encoding
    
    except requests.exceptions.HTTPError as http_err:
        print(f"Loi HTTP da xay ra: {http_err} - Status code: {response.status_code if 'response' in locals() else 'N/A'}") 
        
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Loi ket noi da xay ra: {conn_err}")
        
    except requests.exceptions.Timeout as timeout_err:
        print(f"Loi timeout da xay ra: {timeout_err}")
    
    except requests.exceptions.RequestException as req_err:
        print(f"Loi khong xac dinh da xay ra: {req_err}")
    
    return None  # Nếu có lỗi, trả về None để báo hiệu không có nội dung
```

```python
# Gọi hàm để lấy nội dung trang
html_content = fetch_page_content(TARGET_URL)

if html_content:
    print(f"Đã lấy được nội d ung HTML (kích thước: {len(html_content)} bytes)")
    #Tạm thời lwuu ndung HTML ra file để kiểm tra
    with open("topcv_page_content.html", "wb") as file: # Mở ở chế độ 'wb' (write bytes)
        file.write(html_content)
    print("Nội dung HTML đã được lưu vào file 'topcv_page_content.html' để kiểm tra")
else:
    print("Không thể lấy nội dung trang. Vui lòng kiểm tra lỗi và thử lại")
```
